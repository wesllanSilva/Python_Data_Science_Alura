EXERCICIO:
Sabendo disso, importe os métodos randrange e seed, crie 8 notas de 0 a 10, armazenando seu conteúdo numa variável chamada notas_matematica. Como os números gerados são chamados de pseudo aleatórios, utilize o método seed escolhendo um valor de referência.

*importar os métodos randrange e seed do módulo random*

```python
from random import randrange, seed;
```

*Atribuir um valor inicial através da função seed, para gerar a mesma sequência de valores pseudo aleatórios*

```python
seed(11)
```

*testar o valor da nota entre 0 e 10*

```python
randrange(0,11)
```

*Criar a lista chamada notas_matematica e criar um loop for atribuindo a lista com o append, o valor da nota de 0 a 10 - precisa ter 8 notas*

```python
notas_matematica = []

#usando o seed para garantir que as notas sempre se repetirem
seed(11)

for notas in range(8):
  notas_matematica.append(randrange(0,11))

#verificando o tamanho da lista
print(len(notas_matematica))

notas_matematica
```

#Plotando com Matplotlib

Importando o **matplotlib**

```python
import matplotlib.pyplot as plt
```

```python
x = list(range(1, 9))
y = notas_matematica

plt.plot(x, y, marker='o')

plt.title('Notas de matemática')
plt.xlabel('Provas')
plt.ylabel('Notas')
plt.show()
 
```
